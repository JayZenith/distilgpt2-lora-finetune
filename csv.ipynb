```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv("arxiv_dataset.csv")

# Look at column names 
print(df.columns)

# Quick look
display(df.head())
display(df.info())

# Count papers per category
plt.figure(figsize=(10,6)) 
sns.countplot(data=df, x='Category', order=df['Category'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Number of Papers per Category")
plt.show()

# Trend over years (if 'published' column exists)
if 'published' in df.columns:
    df['published'] = pd.to_datetime(df['published'])
    yearly_counts = df.groupby(df['published'].dt.year).size()
    yearly_counts.plot(kind='line', marker='o', figsize=(10,5), title='Papers Published per Year')
    plt.ylabel("Count")
    plt.show()
```
